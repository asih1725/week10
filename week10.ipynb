```python
#Recursion Example
def tri_recursion(k):
  if(k > 0):
    result = k + tri_recursion(k - 1)
    print(result)
  else:
    result = 0
  return result

print("\n\nRecursion Example Results")
tri_recursion(6)
```

```python
#Weekly Exercise 2
import math

def luas_lingkaran(jari_jari):
  """menghitung luas lingkaran

  args:
  jari_jari: jari-jari lingkaran.

  returns:
  luas lingkaran.
  """
  return math.pi * jari_jari ** 2

def luas_persegi(sisi):
  """menghitung luas persegi

  args:
  sisi:panjang sisi persegi.

  returns:
  luas persegi.
  """
  return sisi ** 2


def luas_segitiga(alas, tinggi):
  """menghitung luas segitiga

  args:
   alas: panjang alas segitiga.
   tinggi:tinggi segitiga.

  returns:
   luas segitiga.
   """
  return 0.5 * alas * tinggi

while True:
  print("1. luas lingkaran")
  print("2. luas persegi")
  print("3. luas segitiga")
  print("4. selesai")
  pilihan = input("pilih 1, 2, 3, atau 4: ")

  if pilihan == "1":
    jari_jari = float(input("masukkan jari-jari lingkaran: "))
    print("luas lingkaran adalah", luas_lingkaran(jari_jari))
  elif pilihan == "2":
    sisi = float(input("masukkan panjang sisi persegi: "))
    print("luas persegi adalah", luas_persegi(sisi))
  elif pilihan == "3":
    alas = float(input("masukkan panjang alas segitiga: "))
    tinggi = float(input("masukkan tinggi segitiga: "))
    print("luas segitiga adalah", luas_segitiga(alas, tinggi))
  elif pilihan == "4":
    break
  else:
    print("pilihan tidak valid")

print("program selesai")
```
